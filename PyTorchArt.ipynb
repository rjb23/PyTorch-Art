```python
import numpy as np
from PIL import Image

# Set the image size
image_size = 320

# Create an array of zeros with shape (image_size, image_size) for the grayscale image
image = np.zeros((image_size, image_size), dtype=np.uint8)

# Set every other pixel to 255 to create a checkerboard pattern
for i in range(image_size):
    for j in range(image_size):
        if (i+j) % 2 == 0:
            image[i, j] = 255

# Convert the numpy array to a PIL Image
pil_image = Image.fromarray(image)

# Show the image
pil_image.show()
```

```python
import random
from PIL import Image

# Define the size of the image
image_size = 500

# Create a new image of size (image_size, image_size)
image = Image.new('RGB', (image_size, image_size))

# Loop through each pixel in the image
for x in range(image_size):
    for y in range(image_size):
        # Generate a random color
        r = random.randint(0, 255)
        g = random.randint(0, 255)
        b = random.randint(0, 255)
        
        # Set the color of the pixel
        image.putpixel((x, y), (r, g, b))

# Save the image as a PNG file
image.save('random_color_image.png')

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image from the file
img = mpimg.imread('random_color_image.png')

# Display the image
plt.imshow(img)
plt.show()
```
